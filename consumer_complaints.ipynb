## 1. What are the top 10 complaints against products and sub-products?


```python
df_original = pd.read_csv("Consumer_Complaints.csv")
df_original.head()
```

```python
len(df_original.Issue.unique())
```

```python
df = df_original.copy()
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["product_frequency"] = 0
```

```python
df.head()
```

```python
df.groupby("Product").count().sort_values(by="product_frequency", ascending=False)
```

```python
# Finding top 10 complaints against products

df.groupby("Product").count().sort_values(by="product_frequency", ascending=False).iloc[0:10, -1]
```

```python
len(df['Sub-product'].unique())
```

```python
len(df.Product.unique())
```

```python
# Finding top 10 complaints against sub-products

df.groupby(["Product", "Sub-product"]).count().sort_values(by="product_frequency", ascending=False).iloc[0:10, -1]
```

## 2. What are the top 20 issues and sub issues?

```python
# Finding top 20 Issues

df.groupby("Issue").count().sort_values(by="product_frequency", ascending=False).iloc[0:20, -1]
```

```python
# Finding top 20 sub-issues

df.groupby(["Issue", "Sub-issue"]).count().sort_values(by="product_frequency", ascending=False).iloc[0:20, -1]
```

## 3. In which month did most complaints occur and on which day of the week are most complaint received?

```python
df['Date received']  = pd.to_datetime(df['Date received'])
```

```python
df['Year']  = df['Date received'].dt.year
```

```python
df['Month']  = df['Date received'].dt.month_name()
```

```python
df['Day']  = df['Date received'].dt.day_name()
```

```python
df['Day'].value_counts()
```

```python
df["Year"].value_counts()
```

```python
df["Month"].value_counts()
```

## 4. Which state received the largest number of complaints?
